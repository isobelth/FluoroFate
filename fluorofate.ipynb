# FluoroFate

Napari-based GUI for fluorescence cell-death / cell-fate analysis.  
See [FluoroFate_README.md](FluoroFate_README.md) for full documentation.

## Pipeline
1. **Cellpose** segments brightfield frames into cell masks
2. **TrackMate** links cell identities across frames (headless via PyImageJ)
3. **Fluorescence thresholding** identifies positive regions per fluorophore (up to 3)
4. **Fate assignment** classifies cells in one of two modes:
   - **Persistent** — first fluorophore to appear determines permanent fate (e.g. Annexin V / PI)
   - **Snapshot** — per-frame classification by current active fluorophores (e.g. FUCCI)
5. **Outputs** — percentage trajectory plots, spatial trajectory plots, and cell-level CSVs (with positive area per fluorophore) at multiple frame-presence cutoffs
